```python
#Smart Light Controller
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def set_action(self, action):
        # Convert action to uppercase to ensure consistency
        self.status = action.upper()

    def display_status(self):
        print(f"{self.name} is {self.status}")

# Input section
name = input("Light Name: ")
action = input("Action: ")

# Execution
light = SmartLight(name)
light.set_action(action)
light.display_status()
```

```python
# Employee id card system
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        print("\nEmployee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# Input section
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")

# Execution
emp = Employee(name, emp_id, dept)
emp.display_card()
```

```python
#mobile contact record
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("\nContact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

# Input section
c_name = input("Contact Name: ")
c_phone = input("Phone Number: ")

# Execution
contact = Contact(c_name, c_phone)
contact.display_contact()
```

```python
# product price tag generator
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_tag(self):
        print(f"\nProduct: {self.name}")
        print(f"Price: ₹{self.price}")

# Input section
p_name = input("Product Name: ")
p_price = input("Price: ")

# Execution
prod = Product(p_name, p_price)
prod.print_tag()
```

```python
# Movie rating display system
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print(f"\nMovie: {self.name}")
        print(f"Rating: {self.rating} / 5")

# Input section
m_name = input("Movie Name: ")
m_rating = input("Rating: ")

# Execution
movie = Movie(m_name, m_rating)
movie.display_details()
```

```python
# Delivery address manager
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display_delivery(self):
        print("\nDelivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")

# Input section
cust_name = input("Customer Name: ")
cust_addr = input("Address: ")

# Execution
delivery = Delivery(cust_name, cust_addr)
delivery.display_delivery()
```
