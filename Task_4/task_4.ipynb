```python
# Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    
    while True:
        try:
            amount = int(input("Enter recharge amount: ₹"))
            if amount < 50:
                print("Error: Minimum recharge amount is ₹50. Please try again.")
                continue
            
            if amount not in valid_plans:
                print(f"Error: ₹{amount} is not a valid plan. Valid plans are: {valid_plans}")
                continue
            
            print(f"Success! Recharge of ₹{amount} is successful.")
            break # Exit loop on success
        except ValueError:
            print("Invalid input. Please enter a numerical value.")

validate_recharge()
```

```python
#Inventory Reorder Alert System
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(f"Reorder Alert: {item} (Stock: {stock})")
        else:
            print(f"Stock OK: {item} (Stock: {stock})")

# Example Data
inventory = {"Apples": 20, "Bananas": 10, "Oranges": 15, "Milk": 5}
check_inventory(inventory)
```

```python
#Student Result Processing System
def process_student_result(marks_list):
    total = 0
    for mark in marks_list:
        total += mark
    
    average = total / len(marks_list)
    print(f"Average Marks: {average:.2f}")
    
    return "Pass" if average >= 50 else "Fail"

# Example Usage
student_marks = [55, 60, 45, 70]
result = process_student_result(student_marks)
print(f"Final Result: {result}")
```

```python
#Cab Fare Estimator with Retry Option
def calculate_fare(km, is_peak):
    base_fare = 50
    rate_per_km = 12
    fare = base_fare + (km * rate_per_km)
    
    if is_peak:
        fare *= 1.25  # Adding 25% extra
    return fare

def fare_system():
    while True:
        try:
            distance = float(input("Enter distance in km: "))
            peak_input = input("Is it peak hour? (yes/no): ").lower()
            peak_status = True if peak_input == 'yes' else False
            
            total_fare = calculate_fare(distance, peak_status)
            print(f"Estimated Fare: ₹{total_fare:.2f}")
            
            retry = input("Do you want to calculate another fare? (yes/no): ").lower()
            if retry != 'yes':
                break
        except ValueError:
            print("Invalid input. Please enter numbers for distance.")

fare_system()
```

```python
#Employee Attendance Eligibility
def check_attendance_eligibility(attendance_list):
    present_days = 0
    for status in attendance_list:
        if status == "P":
            present_days += 1
            
    percentage = (present_days / len(attendance_list)) * 100
    print(f"Attendance: {percentage:.2f}%")
    
    return "Eligible" if percentage >= 75 else "Not Eligible"

# Example Usage
attendance = ["P", "P", "A", "P", "P", "P", "A", "P"]
print(f"Status: {check_attendance_eligibility(attendance)}")
```

```python
#Password Strength Checker
def check_password_strength(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False
    
    if len(password) < 8:
        return "Weak: Minimum 8 characters required."
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
            
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak: Must contain a digit and a special character (@, #, or $)."

# Example Usage
pwd = input("Enter password to check: ")
print(check_password_strength(pwd))
```
