```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_engagement(likes_list):
    # Rule: Calculate total likes using a loop or sum function
    total_likes = 0
    for like in likes_list:
        total_likes += like
    
    # Rule: If Total Likes >= 1000 -> Viral Post, otherwise Normal
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# Example Test
analyze_engagement([500, 200, 350])
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System
def monitor_medicine_stock(medicine_name, stock_level):
    # Rule: Stock < 10 -> Low Stock Alert, else Stock Sufficient
    if stock_level < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    
    print(f"Medicine: {medicine_name}")
    print(f"Medicine Stock: {stock_level}")
    print(f"Status: {status}")

# Example Test
monitor_medicine_stock("Aspirin", 6)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
def check_rainfall_adequacy(rainfall_data, required_level):
    # Rule: Calculate average rainfall using loop and math
    total_rainfall = sum(rainfall_data)
    average_rainfall = total_rainfall / len(rainfall_data)
    
    # Rule: Compare average to required level
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print(f"Average Rainfall: {average_rainfall}")
    print(f"Rainfall Status: {status}")

# Example Test
check_rainfall_adequacy([60, 80, 75, 73], 70)
```

```python
#Problem 4: Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
    # Rule: A set only stores unique values. 
    # If the set length is shorter than the list, duplicates exist.
    if len(set(usernames)) < len(usernames):
        result = "Yes"
    else:
        result = "No"
    
    print(f"Duplicate Accounts Found: {result}")

# Example Test
detect_duplicates(["alex_99", "sam_01", "alex_99", "nina_v"])
```

```python
# Problem 5: Healthcare – Appointment Eligibility Checker
def check_appointment_eligibility(age):
    # Rule: Age >= 18 is eligible
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}")

# Example Test
check_appointment_eligibility(21)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
def filter_premium_crops(price_list):
    # Rule: Filter prices > 2000
    premium_crops = [price for price in price_list if price > 2000]
    
    print(f"Premium Crops: {premium_crops}")

# Example Test
filter_premium_crops([1500, 2500, 3200, 1900])
```

```python
#Problem 7: System Monitoring – Application Health Checker
def app_health_checker(error_count):
    # Rule: Conditional logic for health status
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

# Example Test
app_health_checker(7)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
def validate_transaction(amount):
    limit = 50000
    
    # Rule: Check if amount is within limit
    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

# Example Test
validate_transaction(60000)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility
def check_exam_eligibility(attendance_list):
    # Attendance list: 1 for Present, 0 for Absent
    # Rule: Attendance % = (Days Present / Total Days) * 100
    percentage = (sum(attendance_list) / len(attendance_list)) * 100
    
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {status}")

# Example Test
check_exam_eligibility([1, 1, 1, 0, 1]) # 4 out of 5 is 80%
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def calculate_electricity_bill(units):
    bill = 0
    
    # Billing Rules (Slab Logic)
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        # First 100 at ₹3, next at ₹5
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        # First 100 at ₹3, second 100 at ₹5, remainder at ₹7
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
        
    print(f"Units Consumed: {units}")
    print(f"Total Bill: ₹{bill}")
    print(f"Usage Level: {usage}")
    return bill, usage

# Example Test
calculate_electricity_bill(250)
```
