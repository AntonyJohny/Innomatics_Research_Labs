```python
#User Login check
correct_username = "admin"
correct_password = "1234"

# Manually taking input from the user
username_input = input("Enter username: ")
password_input = input("Enter password: ")

# Checking if both match
if username_input == correct_username and password_input == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#Pass/fail analyzer
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0

for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

print(f"Total Pass: {pass_count}")
print(f"Total Fail: {fail_count}")
```

```python
#Simple Data cleaner
names = [" Alice ", "bob", " CHARLIE "]
# Using list comprehension for a clean solution
cleaned_names = [name.strip().lower() for name in names]
print(cleaned_names)
```

```python
#Message Length analyzer
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    flag = " - (Longer than 10 chars)" if length > 10 else ""
    print(f"Message: '{msg}' | Length: {length}{flag}")
```

```python
#Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = logs.count("ERROR")
print(f"Total Error Count: {error_count}")
```
