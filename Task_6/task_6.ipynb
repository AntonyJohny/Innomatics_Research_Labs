```python
#Smart Parking Lot Management System
def manage_parking(capacity, logs):
    """
    Calculates current occupancy and status of a parking lot.
    """
    count = 0
    for log in logs:
        if log == "IN":
            count += 1
        elif log == "OUT":
            count -= 1
    
    status = "Available" if count < capacity else "Full"
    
    # Alert if capacity is exceeded
    if count > capacity:
        print("ALERT: Capacity Exceeded!")
    
    return count, status

# Sample Execution
cap = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parked, status = manage_parking(cap, logs)

print(f"Currently Parked Vehicles: {parked}")
print(f"Parking Status: {status}")
```

```python
#Online Food Delivery Time Estimator
def estimate_delivery(distance, traffic, weather):
    """
    Calculates ETA based on distance and environmental factors.
    """
    # Base time: 5 mins per km
    base_time = distance * 5
    
    # Dynamic Delays
    traffic_delay = 20 if traffic == "High" else 10 if traffic == "Medium" else 0
    weather_delay = 15 if weather == "Rainy" else 0
    
    total_eta = base_time + traffic_delay + weather_delay
    return total_eta

# Sample Execution
eta = estimate_delivery(8, "High", "Rainy")
print(f"Estimated Delivery Time: {eta} minutes")
```

```python
#Movie Theatre Seat Occupancy Analyzer
def analyze_occupancy(total_seats, booked_list):
    """
    Analyzes booking data to determine theater status.
    """
    booked_count = len(booked_list)
    occupancy_pct = (booked_count / total_seats) * 100
    
    if occupancy_pct == 100:
        status = "Housefull"
    elif occupancy_pct >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"
        
    return occupancy_pct, status

# Sample Execution (150 entries in list)
booked_seats = [1] * 150 
pct, status = analyze_occupancy(200, booked_seats)

print(f"Occupancy: {pct}%")
print(f"Show Status: {status}")
```

```python
#Cloud Server Load Classification System
def classify_server_load(readings):
    """
    Calculates average CPU usage and classifies health status.
    """
    avg_load = sum(readings) / len(readings)
    
    if avg_load < 50:
        status = "Normal"
    elif 50 <= avg_load <= 80:
        status = "Warning"
    else:
        status = "Critical"
        
    return avg_load, status

# Sample Execution
cpu_data = [45, 60, 70, 85, 90]
avg, status = classify_server_load(cpu_data)

print(f"Average CPU Load: {avg}%")
print(f"Server Status: {status}")
```

```python
#Smart Classroom Resource Usage Monitor
def monitor_resources(usage_dict):
    """
    Identifies resources used more than 8 hours and triggers energy alerts.
    """
    overused = [res for res, hours in usage_dict.items() if hours > 8]
    alert = "Yes" if overused else "No"
    
    return ", ".join(overused), alert

# Sample Execution
usage = {"Projector": 6, "AC": 9, "Lights": 4}
overused_res, energy_alert = monitor_resources(usage)

print(f"Overused Resources: {overused_res}")
print(f"Energy Alert: {energy_alert}")
```

```python
#Online Event Registration Capacity Controller
def manage_registration(capacity, registrations):
    """
    Manages event entry and calculates waitlist.
    """
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
        
    return confirmed, waitlisted, status

# Sample Execution
conf, wait, stat = manage_registration(100, 105)

print(f"Confirmed Registrations: {conf}")
print(f"Waitlisted Users: {wait}")
print(f"Registration Status: {stat}")
```
