```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest score
max_score = max(employees.values())

# Identify all employees with that score
top_performers = [name for name, score in employees.items() if score == max_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

```python
query = "Buy mobile phone buy phone online"

# Clean and split the data
words = query.lower().split()
word_counts = {}

for word in words:
    word_counts[word] = word_counts.get(word, 0) + 1

# Filter for keywords appearing more than once
duplicates = {word: count for word, count in word_counts.items() if count > 1}

print(duplicates)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Identify even readings and store as (hour, value)
valid_readings = [(idx, val) for idx, val in enumerate(sensor_readings) if val % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_counts = {}
total_emails = len(emails)

# Extract domain and count
for email in emails:
    domain = email.split('@')[1]
    domain_counts[domain] = domain_counts.get(domain, 0) + 1

# Calculate and display percentages
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
avg_sales = sum(sales) / len(sales)
threshold = avg_sales * 1.3

# Detect spikes (Day index starts at 1)
for i, sale in enumerate(sales):
    if sale > threshold:
        print(f"Day {i+1}: {sale}")
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

counts = {}
for uid in user_ids:
    counts[uid] = counts.get(uid, 0) + 1

# Display duplicates
for uid, count in counts.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
